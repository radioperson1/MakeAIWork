```python
# Bepaal en importeer benodigde librairies

import json 
import pandas as pd
import requests
# Youtube Python Request Module: Just the Basics 
import numpy as np
import seaborn as sns
import pprint
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(color_codes=True)
```

```python
# Stap 1 data inlezen 

# Maak en request. Deze url werkt alleen als de rest server gestart is. Dit doe ik via de terminal van git bash door in de juiste directory te gaan staan en dan de server op te starten met ./start_"input".sh

# De parameters van de website zijn "page" en "1", ik verwacht dus dat er meerdere pagina's zijn. Er zijn er in totaal 82, deze data wil ik natuurlijk allemaal gebruiken. Daarom kies ik ervoor om de parameters apart te benoemen ipv de url direct in de response variabele te plakken. Mijn doel is om een manier te verzinnen om er doorheen te loopen.

# 1 > pagenumber = input

url = "http://localhost:8080/medish_centrum_randstad/api/netlify"

# params = {'page':pageno} for loop?

params = {'page': 1}

response = requests.get(url, params)

# Om te checken of het inlezen gelukt is. Status code 200 betekent dat alles goed ging en er geen problemen zijn. Ik voeg een codeblock toe in t geval dat het niet goed is.
# How to Access Web APIs using Python Requests and JSON

response.headers['content-type']

if response.status_code == 200:
    df = pd.read_json(response.text)
else:
    print(f'Error: {response.status_code}')

df.head()
```

```python
# Nu wil ik de laatste kolom 'data' van df extraheren. Dat doe ik door de kolom op te slaan in een variabele.
## Zo te zien hebben we alle data te pakken dus hoeft er geen loop gemaakt te worden om over de pagina's heen te gaan. 

data = df['data']

# Nu wil ik van deze variable (momenteel een dict) een dataframe maken. Daarvoor gebruiken een pandas trucje

df = data.to_frame()
df.head()
# df = pd.DataFrame(data)
# display(df)
```
